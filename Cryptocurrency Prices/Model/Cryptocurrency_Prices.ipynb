```python
def plot_df(df, x, y, title="", xlabel='Date', ylabel='Marketcap', dpi=100):
    plt.figure(figsize=(20,8), dpi=dpi)
    plt.plot(x, y, color='tab:red')
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.show()


plot_df(df, x=df['Date'], y=df['Marketcap'], title='Marketcap of AAVE according to time')
```
